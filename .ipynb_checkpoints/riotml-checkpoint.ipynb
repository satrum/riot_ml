```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(results, y, test_size=0.5)
print (X_train.shape, y_train.shape)
print (X_test.shape, y_test.shape)
```

```python
baseline_model = keras.Sequential([
    # `input_shape` is only required here so that `.summary` works. 
    keras.layers.Dense(16, activation=tf.nn.relu, input_shape=(282,)),
    keras.layers.Dense(16, activation=tf.nn.relu),
    keras.layers.Dense(1, activation=tf.nn.sigmoid)
])

baseline_model.compile(optimizer='adam',
                       loss='binary_crossentropy',
                       metrics=['accuracy', 'binary_crossentropy'])

baseline_model.summary()
```

```python
baseline_history = baseline_model.fit(X_train,
                                      y_train,
                                      epochs=50,
                                      batch_size=512,
                                      validation_data=(X_test, y_test),
                                      verbose=0)
```

```python
l2_model = keras.models.Sequential([
    keras.layers.Dense(16, kernel_regularizer=keras.regularizers.l2(0.001),
                       activation=tf.nn.relu, input_shape=(282,)),
    keras.layers.Dense(16, kernel_regularizer=keras.regularizers.l2(0.001),
                       activation=tf.nn.relu),
    keras.layers.Dense(1, activation=tf.nn.sigmoid)
])

l2_model.compile(optimizer='adam',
                 loss='binary_crossentropy',
                 metrics=['accuracy', 'binary_crossentropy'])

baseline_model.summary()
```

```python
l2_model_history = l2_model.fit(X_train, y_train,
                                epochs=50,
                                batch_size=512,
                                validation_data=(X_test, y_test),
                                verbose=0)
```

```python
dpt_bigger_model = keras.models.Sequential([
    keras.layers.Dense(128, activation=tf.nn.relu, input_shape=(282,)),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(128, activation=tf.nn.relu),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation=tf.nn.sigmoid)
])

dpt_bigger_model.compile(optimizer='adam',
                 loss='binary_crossentropy',
                 metrics=['accuracy', 'binary_crossentropy'])

dpt_bigger_model.summary()
```

```python
dpt_bigger_history = dpt_bigger_model.fit(X_train, y_train,
                                epochs=50,
                                batch_size=2048,
                                validation_data=(X_test, y_test),
                                verbose=0)
```

```python
def plot_history(histories, key='binary_crossentropy'): # , acc
  plt.figure(figsize=(16,10))
    
  for name, history in histories:
    val = plt.plot(history.epoch, history.history['val_'+key],
                   '--', label=name.title()+' Val')
    plt.plot(history.epoch, history.history[key], color=val[0].get_color(),
             label=name.title()+' Train')

  plt.xlabel('Epochs')
  plt.ylabel(key.replace('_',' ').title())
  plt.legend()

  plt.xlim([0,max(history.epoch)])

plot_history([('l2_bigger', l2_bigger_history)], key='binary_crossentropy')
plot_history([('l2_bigger', l2_bigger_history)], key='acc')
'''
plot_history([('baseline', baseline_history),('l2', l2_model_history),('l2_bigger', l2_bigger_history),
              ('dpt_bigger', dpt_bigger_history)])

plot_history([('baseline', baseline_history),('l2', l2_model_history),('l2_bigger', l2_bigger_history),
              ('dpt_bigger', dpt_bigger_history)], key='acc')
'''
plt.show()
```

```python
balance = 100
bet = 0.05 #%
win = 2 #win reward multiplied
predict  = l2_bigger_model.predict(X_test,verbose=1)
print(len(predict), len(y_test))
count = 0
for i in range(len(predict)):
    #print(predict[i], y_test.iloc[i], abs(y_test.iloc[i]-predict[i]))
    if (abs(y_test.iloc[i]-predict[i])<0.5).bool():
        count+=1
print(count, count/len(y_test))

for i in range(len(predict)):
    if (abs(0.5-predict[i])>0.14):
        cash = balance * bet
        balance-=cash
    else:
        continue
    if (abs(y_test.iloc[i]-predict[i])<0.5).bool():
        reward = win * cash
        balance +=reward
    print(balance)
print(balance)
```

```python
y1d = y.values.ravel()
print(y1d)
print(y1d.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(results, y1d, test_size=0.25)
print (X_train.shape, y_train.shape)
print (X_test.shape, y_test.shape)

# import machine learning algorithms
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc

#See also
#sklearn.tree.DecisionTreeClassifier, RandomForestClassifier, AdaBoostClassifier
```

```python
# train with Gradient Boosting algorithm
# compute the accuracy scores on train and validation sets when training with different learning rates

learning_rates = [i*0.01 for i in range(10,100,5)]
for learning_rate in learning_rates:
    #gb = GradientBoostingClassifier(n_estimators=242, learning_rate = learning_rate, max_features=2, max_depth = 2, random_state = 0)
    gb = GradientBoostingClassifier(n_estimators=320, learning_rate = learning_rate, max_features=2, max_depth = 2, random_state = 0)
    gb.fit(X_train, y_train)
    print("Learning rate: ", learning_rate)
    print("Accuracy score (training): {0:.3f}".format(gb.score(X_train, y_train)))
    print("Accuracy score (validation): {0:.3f}".format(gb.score(X_test, y_test)))
    print()
```

```python
# Output confusion matrix and classification report of Gradient Boosting algorithm on validation set
#loss : {'deviance', 'exponential'}, optional (default='deviance')
gb = GradientBoostingClassifier(n_estimators=320, learning_rate = 0.25, max_features=2, max_depth = 2, random_state = 0)
#n_estimators=282
gb.fit(X_train, y_train)
predictions = gb.predict(X_test)

print("Accuracy score (training): {0:.3f}".format(gb.score(X_train, y_train)))
print("Accuracy score (validation): {0:.3f}".format(gb.score(X_test, y_test)))
print()
    
print("Confusion Matrix:")
print(confusion_matrix(y_test, predictions))
print()
print("Classification Report")
print(classification_report(y_test, predictions))
```

```python
y_scores_gb = gb.decision_function(X_test)
fpr_gb, tpr_gb, _ = roc_curve(y_test, y_scores_gb)
roc_auc_gb = auc(fpr_gb, tpr_gb)

print("Area under ROC curve = {:0.2f}".format(roc_auc_gb))
```
